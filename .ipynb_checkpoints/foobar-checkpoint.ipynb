```python
import pandas as pd
from seaborn import regplot
import matplotlib.pyplot as plt
from IPython.display import Image
launches = pd.read_pickle('launches.pkl');
launches.groupby('Year').size();
launches.groupby('Year').failed.sum();
history = launches.groupby('Year').failed.agg(['count', 'sum']);
history.columns=['Launches', 'Explosions'];
fig, [ax1, ax2] = plt.subplots(1,2, figsize=(14,7));
regplot('Launches', 'Explosions', data=history, fit_reg=False, ax=ax1);
ax1.set_title("Rocket Launches and Explosions", fontdict ={'fontsize':25});
i = plt.imread("./assets/biasVsVarianceImage.png");
ax2.set_title("every bias and variance explanation EVER", fontdict ={'fontsize':15});
ax2.imshow(i);
plt.show()
```
